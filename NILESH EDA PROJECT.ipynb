# Kernal Density Estimatation (KDE) of Penguin Species

```python
sns.FacetGrid(df, hue="Species", size=6,) \
   .map(sns.kdeplot, "Flipper Length (mm)",shade=True) \
   .add_legend()
```

 KDE for DataSet is done for representation of Flipper Length of Species to Distinguish between each other
